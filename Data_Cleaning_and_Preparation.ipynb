# 3 Check for data imbalance
### We will use the column 'Revenue' as it contains the desired output of the customer's activity.
### A value of 'True' means that the visitor made a purchase.

```python
fig, axs = plt.subplots(figsize=(5,5))
g = sns.countplot(x='Revenue', data=shoppers_df, ax=axs)
```

```python
shoppers_df['Revenue'].value_counts()
```

### The data is imbalanced.

### 81.7% = False (did not make a purchase), and 18.3% = True (made a purchase)
